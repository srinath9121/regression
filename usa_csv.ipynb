## Data cleaning

### Subtask:
Clean the data by handling missing values, outliers, and inconsistencies.


**Reasoning**:
The data is a single string in each row separated by semicolons.  We need to properly split the string into columns.



```python
import pandas as pd

try:
    df = pd.read_csv('/content/merged_file.csv', sep=';', encoding='latin-1', header=0)
    #Check for any rows where the number of elements after splitting is not equal to the expected number of columns
    #If found, print an error message and skip those rows
    #This will help to skip rows with corrupted data
    expected_columns = len(df.columns)
    df = df[df.iloc[:, 0].str.count(';') == expected_columns - 1]

    # Split the combined string column into multiple columns
    new_df = df['school'].str.split(';', expand=True)
    new_df.columns = df.columns.values

    # Print the first few rows of the new DataFrame to verify the split
    print(new_df.head())
    print(new_df.shape)

except FileNotFoundError:
    print("Error: File not found at the specified path.")
except pd.errors.ParserError:
    print("Error: Unable to parse the CSV file. Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

**Reasoning**:
The data seems to be correctly split now. The next step is to handle the missing values, outliers and inconsistencies.



```python
import pandas as pd

# Assuming 'new_df' is the DataFrame created in the previous step
# Replace missing values with the mean for numerical columns and the mode for categorical columns

for col in new_df.columns:
    if new_df[col].dtype in ['int64', 'float64']:
        new_df[col] = pd.to_numeric(new_df[col], errors='coerce')
        new_df[col].fillna(new_df[col].mean(), inplace=True)
    else:
        new_df[col] = new_df[col].str.replace('"', '')
        new_df[col].fillna(new_df[col].mode()[0], inplace=True)

#Remove duplicate rows
new_df.drop_duplicates(inplace=True)

# Print some info to verify the cleaning
print(new_df.isnull().sum())
print(new_df.dtypes)
print(new_df.head())
```
